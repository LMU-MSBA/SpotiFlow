```python
import pandas as pd
```

```python
df = pd.read_csv('spotify_data.csv')
```

```python
df.head()
```

```python
#hardcode User ID of 1 for each record
df['User ID'] = 1
```

```python
df.columns
```

```python
DimArtist = df[['Artist Name']].drop_duplicates()
```

```python
DimTrack = df[['Track ID', 'Track Name', 'danceability',
       'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature']].drop_duplicates()
```

```python
DimDate = df[['Day_Played']].drop_duplicates()
```

```python
DimUser = df[['User ID']].drop_duplicates()
```

```python
FactUser = df[['id','User ID', 'Track ID', 'Artist Name', 'Day_Played']]
```

```python
FactUser
```

```python
# send all tables to database
DimArtist.to_sql('DimArtist', engine, if_exists='append', index=False)
DimTrack.to_sql('DimTrack', engine, if_exists='append', index=False)
DimDate.to_sql('DimDate', engine, if_exists='append', index=False)
DimUser.to_sql('DimUser', engine, if_exists='append', index=False)
FactUser.to_sql('FactUser', engine, if_exists='append', index=False)
```
